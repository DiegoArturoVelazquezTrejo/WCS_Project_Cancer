```python
import pandas as pd 

data = pd.read_csv("data/counts_normalized_log2_shortlabels.csv")
enoch_shared  = pd.read_csv("Overlap_output.tsv", sep="\t")["Gene"].to_list()
genes  = pd.read_csv("enoch_total_genes.tsv", sep="\t")["Gene"].to_list()

documento = ""
# Abrimos el archivo llamado regulatoryGraph.txt
with open("BooleanModelCoLoMoTo/regulatoryGraph.txt", "r") as file:
    lines = file.readlines()
    # Concatenamos todas las líneas en una sola cadena de texto
    documento = "".join(lines)
```

```python
# Iteramos sobre cada gen y si aparece en el documento, lo guardamos 
genes_in_model = []
for gene in genes:
    if gene in documento:
        genes_in_model.append(gene)

print("Número de genes en el modelo:", len(genes_in_model))
```
